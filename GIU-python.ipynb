```python
def Find_CrossOverFreq(num):
    CrossOverFreq=[]
    Array_CrossOver=df[['F Hz','20logAmp','Find Zero Gain']]  #The Array for CrossOver freq Compute index fun
    select_indices = list(np.where(df["Find Zero Gain"] ==num)[0])
    for i in range(len(select_indices)):
        CrossOverFreq.append(Array_CrossOver['F Hz'].iloc[select_indices[i]])
        
    return CrossOverFreq
```

```python
def Find_PhaseMargin(num):
    PhaseMargin=[]
    Array_CrossOver=df[['F Hz','20logAmp','Find Zero Gain']]  #The Array for CrossOver freq Compute index fun
    select_indices = list(np.where(df["Find Zero Gain"] ==num)[0])
    for i in range(len(select_indices)):
        PhaseMargin.append(df['Find Zero Phase'].iloc[select_indices[i]])
        
    return PhaseMargin
    
```

# Compute suggested exact value for inductor  L

```python
#L =(Vin_Max-Vout)/(IL_Ripple/100*Iload)*Vout/Vin_Max*(1/(Fsw*1000))*1000000
#Vin=28.00
#Vin Min=23.00
#Vin Max=32.00
# Vout=12.00
# Fsw=400.00
# Iload=2.00
# Tss=6.0
# I_limit=4.00
# IL Ripple=50
# Min Vin startup=4.50
```

 $$L=(Vin_Max-Vout)/(IL_Ripple/100*Iload)*Vout/Vin_Max*(1/(Fsw*1000))*1000000 $$


```python
Vin_Max=input("Enter Vin_Max Value = ")
Vout=input("Enter Vout Value = ")
IL_Ripple=input("Enter IL_Ripple Value = ")
Iload=input("Enter Iload Value = ")
Fsw=input("Enter Fsw Value = ")
```

```python
numer=(int(Vin_Max)-int(Vout))*(int(Vout)*1000)*100
den=(int(IL_Ripple)*int(Iload)*int(Vin_Max)*int(Fsw))
L =numer/den
print(L ,"UH")
```

# Here I will Compute L value as a simple exampe in GUI 

```python

from tkinter import *
window=Tk()

class MyWindow():

    def __init__(self, win):
        self.lbl10=Label(win, text="# Compute the exact value for inductorL").pack()
        self.lbl1=Label(win, text='Vin_Max')
        self.lbl2=Label(win, text='Vout')
        self.lbl3=Label(win, text=' IL_Ripple')
        self.lbl4=Label(win, text='Fsw')
        self.lbl5=Label(win, text='I load')  

        
        self.t1=Entry()
        self.t2=Entry()
        self.t3=Entry()
        self.t4=Entry()
        self.t5=Entry()
        
        self.t10=Entry()
        
        self.t111=Entry()
    
        self.lbl1.place(x=50, y=50)
        self.t1.place(x=120, y=50)
        self.lbl2.place(x=50, y=100)
        self.t2.place(x=120, y=100)
        self.lbl3.place(x=50, y=150)
        self.t3.place(x=120, y=150)
        self.lbl4.place(x=50, y=200)
        self.t4.place(x=120, y=200)
        self.lbl5.place(x=50, y=250)
        self.t5.place(x=120, y=250)
        
        
        self.b1=Button(win, text='Result(LValue)', command=self.L_Calculate)
        self.b1.place(x=100, y=300)
        self.t111.place(x=200, y=305)
    
        
    def L_Calculate(self):
        self.t111.delete(0, 'end')
        result=self.numer()/self.den()
        self.t111.insert(END, str(result)+"  UH")
        
    def numer (self):
       # numer=(int(Vin_Max)-int(Vout))*(int(Vout)*1000)*100
        self.t111.delete(0, 'end')
        num1=int(self.t1.get())
        num2=int(self.t2.get())
        result1=(num1-num2)*(num2*100000)
        return result1
  
    def den(self):
        #den=(int(IL_Ripple)*int(Iload)*int(Vin_Max)*int(Fsw))
        self.t111.delete(0, 'end')
        num1=int(self.t3.get())
        num2=int(self.t5.get())
        num3=int(self.t1.get())
        num4=int(self.t4.get())
        result2=num1*num2*num3*num4
        return result2


        

MyWindow(window)



window.title('SiC46xmicroBUCK Family External Component Selection Calculator')
window.geometry("500x1000+10+10")
window.mainloop()




```
